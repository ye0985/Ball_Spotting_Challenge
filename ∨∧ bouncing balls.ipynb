# 1. 50 Random ∨-bouncing Balls 

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

# Set the number of trajectories
num_trajectories = 50

# Set the time for motion
time = 10

# Set the number of time steps
num_steps = 101 

# Set the time interval
dt = time/(num_steps-1) 

# Initialize the trajectories array
trajectories = np.zeros((num_trajectories, num_steps, 2))

# Generate random start positions and velocities for each trajectory
start_positions = np.random.uniform(low=0, high=1, size=(num_trajectories, 2))
velocities = np.random.uniform(low=-2, high=0, size=(num_trajectories, 2))

# Generate the trajectories
for i in range(num_trajectories):
    for j in range(num_steps):
        # Update the position using the velocity
        trajectories[i,j] = start_positions[i] + velocities[i]*dt*j
        
        # Check if the position is below the ground
        if trajectories[i,j,1] < 0:
            # Reflect the trajectory and reduce the velocity
            trajectories[i,j,1] = -trajectories[i,j,1]
    
    
    acceleration = np.diff(trajectories[i, :, :], axis=0) / dt
    mean_acceleration = np.mean(acceleration, axis=0)
    print(f"id가 {i}인 물체의 평균 가속도:", mean_acceleration)

# Store the position data in a CSV file
filename = 'v_trajectories.csv'
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['time', 'id', 'x', 'y'])
    for i in range(num_trajectories):
        for j in range(num_steps):
            writer.writerow([dt*j, i, trajectories[i,j,0], trajectories[i,j,1]])

# Plot the trajectories
plt.figure(figsize=(10,6))
for i in range(num_trajectories):
    plt.plot(trajectories[i,:,0], trajectories[i,:,1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('50 Random V-Motion Bouncing Trajectories')
plt.show()

```

# 2. 50 Random ∧-bouncing Balls

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of trajectories
num_trajectories = 50

# Set the time for motion
time = 10

# Set the number of time steps
num_steps = 101

# Set the time interval
dt = time/(num_steps-1)

# Initialize the trajectories array
trajectories = np.zeros((num_trajectories, num_steps, 2))

# Generate random start positions and velocities for each trajectory
start_positions = np.random.uniform(low=[0, 0], high=[1, 0.5], size=(num_trajectories, 2))
velocities = np.random.uniform(low=[1, 5], high=[2, 10], size=(num_trajectories, 2))

# Generate the trajectories
for i in range(num_trajectories):
    for j in range(num_steps):
        # Update the position using the velocity
        trajectories[i,j] = start_positions[i] + velocities[i]*dt*j
        
        # Check if the trajectory hits the wall
        if trajectories[i,j,0] > 1:
            # Reflect the trajectory and reduce the velocity
            trajectories[i,j,0] = 2 - trajectories[i,j,0]
  

        
        # Update the position using the velocity
        trajectories[i,j] = trajectories[i,j] + [0, -9.81*(dt**2)/2]
        
        # Check if the trajectory hits the floor
        if trajectories[i,j,1] < 0:
            # Reflect the trajectory and reduce the velocity
            trajectories[i,j,1] = -trajectories[i,j,1]
        
    acceleration = np.diff(trajectories[i, :, :], axis=0) / dt
    mean_acceleration = np.mean(acceleration, axis=0)
    print(f"id가 {i}인 물체의 평균 가속도:", mean_acceleration)
    
# Store the position data in a CSV file
filename = '∧_trajectories.csv'
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['time', 'id', 'x', 'y'])
    for i in range(num_trajectories):
        for j in range(num_steps):
            writer.writerow([dt*j, i, trajectories[i,j,0], trajectories[i,j,1]])

# Plot the trajectories
plt.figure(figsize=(10,6))
for i in range(num_trajectories):
    plt.plot(trajectories[i,:,0], trajectories[i,:,1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('50 Random ∧-Motion Bouncing Trajectories')
plt.show()
```
